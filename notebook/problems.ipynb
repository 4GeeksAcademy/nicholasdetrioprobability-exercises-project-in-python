## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np

def dice_roll(n_dice_rolls = 1000):
    count_even_or_greater_than_7 = 0

    for i in range(n_dice_rolls):

        #roll a random number for each dice and store in dice1 or dice 2 value
        total_of_dice = random.randint(1,6) + random.randint(1,6)
        
        if (total_of_dice % 2 == 0) or (total_of_dice > 7):
            count_even_or_greater_than_7 = count_even_or_greater_than_7+1

    return count_even_or_greater_than_7
print("The propability of rolling 2 dice and the sum of the dice being even or greater than 7 is")
print(dice_roll()/1000)



```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random
import numpy as np

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#print(ball_box)


def white_red(ball_pull = 1000):

    white_and_red_count = 0

    for i in range (ball_pull):
        red_count = 0
        white_count = 0

        for i in range (5):

            colors = random.randint (0,59)
            #print(colors)

            if colors < 10:
                white_count=white_count+1
                #print(f"White Count:{white_count}")

            if (colors > 9) and (colors < 30):
                red_count = red_count+1
                #print(f"Red Count:{red_count}")

            if (red_count == 2) and (white_count == 3):
                white_and_red_count = white_and_red_count+1
                #print(f"White and Red Count:{white_and_red_count}")

    return white_and_red_count

def same_color(ball_pull = 1000):

    same_color_count = 0

    for i in range (ball_pull):
        red_count = 0
        white_count = 0
        green_count = 0

        for i in range (5):

            colors = random.randint (0,59)
            #print(colors)

            if colors < 10:
                white_count=white_count+1
                #print(f"White Count:{white_count}")

            if (colors > 9) and (colors < 30):
                red_count = red_count+1
                #print(f"Red Count:{red_count}")

            if (colors >= 30):
                green_count = green_count+1
                #print(f"Green Count:{green_count}")

            if (red_count == 5) or (white_count == 5) or (green_count == 5):
                same_color_count = same_color_count+1
                #print(f"Same Color Count:{same_color_count}")

    return same_color_count


print(f"The probability of 3 white balls and 2 red balls is: {(white_red()/1000)*100}%")
print(f"The probability of 5 balls being the same color is: {(same_color()/1000)*100}%")

            
# TODO
```
